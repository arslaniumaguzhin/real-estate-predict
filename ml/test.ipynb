```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np

x = np.array([[30], [50], [70], [90], [110]])
y = np.array([[3000], [5000], [6500], [9000], [11000]])

model = LinearRegression()
model.fit(x, y)

a = int(input('Введите количество кв.м'))

print(f'Прогноз цены для {a} кв.м:', model.predict([[a]]), 'за 1 кв.м')
```
